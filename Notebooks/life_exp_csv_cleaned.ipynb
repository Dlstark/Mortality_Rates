```python
life_exp[['AVG L/E 1980', 'Male L/E 1980', 'Female L/E 1980']] = life_exp['Life expectancy, 1980*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 1980*', axis=1)
```

```python
life_exp = life_exp.drop(columns=['% Change in Life Expectancy, 1980-2014'])
```

```python
life_exp[['AVG L/E 1985', 'Male L/E 1985', 'Female L/E 1985']] = life_exp['Life expectancy, 1985*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 1985*', axis=1)
life_exp[['AVG L/E 1990', 'Male L/E 1990', 'Female L/E 1990']] = life_exp['Life expectancy, 1990*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 1990*', axis=1)
life_exp[['AVG L/E 1995', 'Male L/E 1995', 'Female L/E 1995']] = life_exp['Life expectancy, 1995*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 1995*', axis=1)
life_exp[['AVG L/E 2000', 'Male L/E 2000', 'Female L/E 2000']] = life_exp['Life expectancy, 2000*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 2000*', axis=1)
life_exp[['AVG L/E 2005', 'Male L/E 2005', 'Female L/E 2005']] = life_exp['Life expectancy, 2005*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 2005*', axis=1)
life_exp[['AVG L/E 2010', 'Male L/E 2010', 'Female L/E 2010']] = life_exp['Life expectancy, 2010*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 2010*', axis=1)
life_exp[['AVG L/E 2014', 'Male L/E 2014', 'Female L/E 2014']] = life_exp['Life expectancy, 2014*'].str.split(expand=True)
life_exp = life_exp.drop('Life expectancy, 2014*', axis=1)
#life_exp[['AVG L/E Change Over Time', 'Male L/E Change Over Time', 'Female L/E Change over Time']] = life_exp['% Change in Life Expectancy, 1980-2014'].str.split(expand=True)
#life_exp = life_exp.drop('% Change in Life Expectancy, 1980-2014', axis=1)

life_exp
```

```python
df_obj = life_exp.select_dtypes(['object'])
life_exp[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
```

```python
life_exp
```

```python
life_exp['Male L/E 1980'] = life_exp["Male L/E 1980"].str.replace('(', '').str.replace(')', '').str.replace(',', '')
life_exp['Female L/E 1980']= life_exp["Female L/E 1980"].str.replace('(', '').str.replace(')', '').str.replace(',', '')
life_exp
```

```python
for col in life_exp.columns:
    life_exp[col] = life_exp[col].str.replace('(', '').str.replace(')', '').str.replace(',', '')
life_exp
```

```python
life_exp_melt = life_exp.melt(id_vars=['Location', 'County'],var_name='Var/Year', value_name='Age in Years')
```

```python
life_exp_melt[['Sex', 'Year']] = life_exp_melt['Var/Year'].str.split(pat = ' L\/E ', expand = True)
```

```python
life_exp_melt
```

```python
life_exp_melt = life_exp_melt.drop(columns=['Var/Year'])
life_exp_melt
```

```python
life_exp_melt['Sex'] = life_exp_melt['Sex'].str.replace('AVG','Combined')
```
